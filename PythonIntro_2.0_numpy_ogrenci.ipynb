Numpy dizi işlemleri için EEM alanında diğer bir örnek uygulama ses işlemedir. Ses sinyalleri tek boyutlu diziler şeklinde okunup işlenebilirler. Altta basit bir sinyal enerjisi hesaplayıp sinyal dalga formu ile beraber çizdirme örneği sunulmuştur

```python
import numpy as np
import matplotlib.pyplot as plt
#.wav dosyası okuyabilmek için pysoundfile kütüphanesini kurup yükleyip kullanalım
!pip install pysoundfile
import soundfile as sf
```

Sinyalleri genelde bölge bölge işleriz. Bu işleme pencereleme (windowing) adı verilir. 

![windowing](https://d3i71xaburhd42.cloudfront.net/1a0f5fd44616ffe227446dc3d22d0bcea1a423df/3-Figure2-1.png)

```python
#Genlik normalizasyonu (senkron çizim yapmayı kolaylaştıracaktır)
x = x / np.max( np.abs(x) )

#Parametreler
pencereUzunlugu = 1024
kaydirmaMiktari = 512

#Pencereler başlangıç endeklerinin belirlenmesi
baslangicNoktalari = np.arange(0, x.size-pencereUzunlugu, kaydirmaMiktari, dtype=int)
pencereSayisi = baslangicNoktalari.size
```

```python
#Sinyali çizdirelim
plt.figure(figsize=(16, 4))
t = np.arange(0,(x.size)/orneklemeFrekansi,1/orneklemeFrekansi)
plt.plot(t,x,label='Orijinal sinyal')
plt.ylabel('Genlik')
plt.xlabel('Zaman(saniye)')
plt.axis('tight')#şekli figür kutusuna oturtacak şekilde ger
plt.legend()
```

N örneğe sahip ayrık zamanlı bir işaretin (x[n]) enerjisi şu şekilde hesaplanabilir:
\begin{equation}
enerji = \sum_{n = 0}^{N-1} |x[n]|^2
\end{equation}


```python
#Her pencere için enerjiyi hesaplayıp çizdirelim
NRG = np.zeros_like(baslangicNoktalari, dtype = float)#enerji değerlerini saklayacağımız dizi: NRG

#Pencereleme işlemi ve enerji hesabı
for k in range(pencereSayisi):
    baslangic = baslangicNoktalari[k]
    x_win = x[baslangic:baslangic + pencereUzunlugu]
    NRG[k] = np.sum(np.power(x_win,2))

#Enerji sinyalini de normalize edelim (senkron çizim yapmayı kolaylaştıracaktır)
NRG = NRG / np.max(NRG)

plt.figure(figsize=(16, 4))
plt.plot(x,label='Orjinal sinyal')
plt.plot(np.arange(NRG.size)*(x.size/NRG.size),NRG,'k',label='Normalize edilmiş enerji')
plt.ylabel('Genlik')
plt.xlabel('Endeks')
plt.axis('tight')
plt.legend(loc='best')
```

**Uygulama:**
Yukarıda ses sinyali ile beraber enerji özniteliğinin çizdirilmesine dair bir örnek sunulmuştur. Buna benzer şekilde hesaplayıp kullanabileceğimiz bir öznitelik sıfır kesim oranıdır. Sinyallerin periyodik olup olmadığının tespitinde yaygın kullanılan bir özniteliktir. Yukarıda enerji için yapılmış olan hesaplama ve çizdirme işlemlerini sıfır kesim oranı için yapan bir kod parçasını bir sonraki hücre içerisine yazınız. 


Sıfır kesim oranı ('Zero crossing rate'):
$$
Sıfır Kesim Oranı = \frac{1}{N-1} \sum_{n=1}^{N-1} | sign(x[n]) - sign(x[n-1]) |
$$
signum fonksiyon tanımı:
$$
sign(x[n])  = \begin{cases}
  1, & \text{ if } x[n] > 0\\ 
  0, & \text{ if } x[n] = 0\\ 
  -1,&\text{ if } x[n] < 0
\end{cases}
$$

Signum fonksiyonu için ***np.sign()*** kullanabilirsiniz
