**Soru-5)** Bu problem için bu klasörde bulunan iki dosya kullanılmaktadır: dog.jpg ve dog_copyright.jpg. Amacınız copyright görüntüsünü dog.jpg görüntüsünün kırmızı renk matrisi içerisine yerleştirmektir. Burada amaç görüntünün başka birisinin farkedemeyeceği şeklide değiştirilmesi ve istenildiğinde bu bilgiye erişilebilmesidir. Bunu gömülecek görüntüdeki değer beyaz olduğunda kırmızı değerleri en yakın çift, siyah olduğund ise tek sayıya eşleyerek yapabiliriz.  Görüntü gömen ve gömülü bilgiyi çeken iki fonksiyon yazarak kullanımı örnekleyiniz. 

```python
img = np.array(Image.open('dog.jpg'))
copyright_img = np.array(Image.open('dog_copyright.jpg'))
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(img);
plt.subplot(1,2,2)
plt.imshow(copyright_img);
# Beklenen çıktı
```

```python
def imaj_gom(orjinal_imaj, gomulecek_imaj):
  yeni_imaj = orjinal_imaj.copy()
  # fonksiyonu tamamlayınız
  return yeni_imaj

# Yukarıdaki fonksiyonu tamamlayınız
yeni_imaj = imaj_gom(img, copyright_img)
plt.imshow(yeni_imaj);
# Beklenen çıktı
```

```python
def gomulu_resmi_cek(orjinal_imaj):
  yeni_imaj = np.zeros_like(orjinal_imaj)
  # fonksiyonu tamamlayınız
  return yeni_imaj.astype('uint8')

# Yukarıdaki fonksiyonu tamamlayınız
gomulu_imaj = gomulu_resmi_cek(yeni_imaj)
plt.imshow(gomulu_imaj);
# Beklenen çıktı
```
